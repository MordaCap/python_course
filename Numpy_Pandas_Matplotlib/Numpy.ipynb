```python
import numpy as np

def make_2_dim_str_np_arr(n):
    return np.array(
        [['{}{}'.format(y,x)
        for x in range(n)]
            for y in range(n)],
        dtype='str')

def make_3_dim_str_np_arr(n):
    return np.array(
        [[['{}{}{}'.format(z,y,x)
        for x in range(n)]
            for y in range(n)]
                for z in range(n)],
        dtype='str')

def make_4_dim_str_np_arr(n):
    return np.array(
        [[[['{}{}{}{}'.format(t,z,y,x)
            for x in range(n)]
                for y in range(n)]
                    for z in range(n)]
                        for t in range(n)],
        dtype='str')
```

```python
import numpy as np
import time

N = 2048

def f(x,y):
    return (x + 1)*(y + 1)

start = time.time()
#a = np.array([[i*j for i in range(1,N+1)] for j in range (1,N+1)], dtype='float64')
a = np.fromfunction(f, (N,N), dtype='float')
#a = np.fromfunction(lambda x, y: (x + 1)*(y + 1), (N,N), dtype='float')
#a = [[i*j for i in range(1,N+1)] for j in range (1,N+1)]
#a = np.array(a, dtype='float64')
#a = np.random.random((N,N))
print ('Matrix generation          {:f} sec'.format(time.time() - start))

start = time.time()
at = a.transpose()
print ('Matrix transpose           {:f} sec'.format(time.time() - start))

start = time.time()
am = at.dot(a)
print ('Matrix multiplication      {:.3f} sec'.format(time.time() - start))
print ('am[234,345] =  {:.2f}'.format(am[345-1,234-1]))
```
